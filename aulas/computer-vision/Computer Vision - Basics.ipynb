# Computer Vision

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('images/teste.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray_img,(5,5),0)
_,bin_img = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

img = img[:,:,::-1] # changing BGR OpenCV format to RGB expected by plt 
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # changing BGR OpenCV format to RGB expected by plt 

cmap_img = {None: img, "gray": gray_img, "binary": bin_img}

fig = plt.figure(figsize=(20,10))
columns = 3
for i, cmap in enumerate(cmap_img):
    ax = plt.subplot(len(cmap_img.items()) / columns + 1, columns, i + 1)
    ax.imshow(cmap_img[cmap], cmap)
    if cmap is None:
        cmap = 'rgb'
    ax.set_title("{} colorspace".format(cmap))
```

```python
vals = bin_img.flatten()
data_hist, bins, patches = plt.hist(vals, bins=2, edgecolor='black')

# bins_label_pos = [(bins.mean()/2), ((bins[2] - bins[1])/2)+bins[1]]
bins_label_pos = [(bins.mean()/2), ((bins[1] + bins[2])/2)]
plt.title("Histogram of {} points".format(len(vals)))
# fig_size = plt.rcParams["figure.figsize"]
# fig_size[0] = 10
# fig_size[1] = 6
# plt.rcParams["figure.figsize"] = fig_size
plt.xticks(bins_label_pos, ['0', '255'])
plt.xlabel("Label")
plt.ylabel("Frequency")
for i in range(2):
    plt.text(bins_label_pos[i] - 15,
             data_hist[i] + 10000,
             str(int(data_hist[i])))
```

### Slicing image

#### Bottom

```python
half = int(bin_img.shape[0]/2)
bot_img = bin_img[half:, :] 
plt.imshow(bot_img, cmap='binary')
print("New shape: {}, Original shape: {}".format(bot_img.shape, bin_img.shape))
```

#### Top

```python
half = int(bin_img.shape[0]/2)
top_img = bin_img[:half, :] 
plt.imshow(top_img, cmap='binary')
print("New shape: {}, Original shape: {}".format(top_img.shape, bin_img.shape))
```

#### Left

```python
half = int(bin_img.shape[1]/2)
left_img = bin_img[:, :half]
plt.imshow(left_img, cmap='binary')
print("New shape: {}, Original shape: {}".format(left_img.shape, bin_img.shape))
```

#### Right

```python
half = int(bin_img.shape[1]/2)
right_img = bin_img[:, half:]
plt.imshow(right_img, cmap='binary')
print("New shape: {}, Original shape: {}".format(right_img.shape, bin_img.shape))
```
